```python
# Install gensim, a popular library for topic modeling and NLP
!pip install gensim
# Install scikit-learn
!pip install scikit-learn

# Import necessary libraries
import gensim.downloader as api # For downloading pre-trained models
from sklearn.decomposition import PCA # For dimensionality reduction
import matplotlib.pyplot as plt # For plotting
import numpy as np # For numerical operations

print("Libraries are ready!")
```

```python
# This will download the model (around 128MB). It might take a minute or two.
print("Loading the 'glove-wiki-gigaword-50' model...")
glove_model = api.load('glove-wiki-gigaword-50')
print("Model loaded successfully!")
```

```python
# Get the vector for the word 'computer'
computer_vector = glove_model['computer']

# Get the vectors for words "processor", "tomato", "apple":
# ...
processor_vector = glove_model['processor']
tomato_vector = glove_model['tomato']
apple_vector = glove_model['apple']
# Print the first 10 numbers of the vector

print(processor_vector[:10])


# Print the total number of dimensions
# Shows up in the model name glove-wiki-gigaword-50
print(len(processor_vector))

# Define a cosine similarity function:
def cosine_similarity(vec1, vec2):
    
  # Formula: dot product of vec1 and vec2 divided by the product of their norms
  # ...
    
    # dot_prod = torch.dot(torch.tensor(vec1), torch.tensor(vec2))
    # norm_prod = torch.torch.linalg.vector_norm()
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)

# Calculate similarities between the vectors
sim_computer_processor = cosine_similarity(computer_vector, processor_vector)
sim_computer_tomato = cosine_similarity(computer_vector, tomato_vector)
sim_computer_apple = cosine_similarity(computer_vector, apple_vector)
sim_apple_tomato = cosine_similarity(apple_vector, tomato_vector)
sim_apple_processor = cosine_similarity(apple_vector, processor_vector)

# Print the cosine similarity scores:
print(f"Computer and processor: {sim_computer_processor:.4f}")
print(f"Computer and tomato: {sim_computer_tomato:.4f}")
print(f"Computer and apple: {sim_computer_apple:.4f}")
print(f"Apple and tomato: {sim_apple_tomato:.4f}")
print(f"Apple and processor: {sim_apple_processor:.4f}")

```

```python
# A curated list of words to see different relationships
#words_to_plot = [
#    'king', 'queen', 'man', 'woman',
#    'france', 'paris', 'germany', 'berlin',
#    'walk', 'walking', 'swim', 'swimming'
#] # Later, consider adding more words

words_to_plot1 = [
    'tomato',
    'processor',
    'apple',
    'computer',
    'laptop',
    'chip',
    'motherboard'
]

# As a matter of fact, I chose this list of words expecting to see Coruna aligned with sea and rain,
# and Madrid with summer and hot.
# That didn't happen and summer and hot were aligned together, and correlated with swim and sea, which
# would also make sense, just not mapping with what I was thinking about.
# I guess the model was trained to match cities, geographical objects and maybe summer activivies? :D

words_to_plot = [
    'madrid',
    'coruna',
    'sea',
    'island',
    'swim',
    'hot',
    'summer',
    'rain',
    'cruisers',
    'capital',
    'galician'
]

# Get the 50-dimensional vector for each word
word_vectors = np.array([glove_model[word] for word in words_to_plot])
# royalty = glove_model.most_similar(positive=['woman', 'king'], negative=['man'])
# comes from the highest number, we do get the king, substract man and add woman, which ofc it's queen xD
# print(f"king - man + woman = {royalty[0][0]}")
```

```python
# Initialize Principle Components Analysis to reduce to 2 components (dimensions)
pca = PCA(n_components=2)

# Fit and transform the word vectors
pca_results = pca.fit_transform(word_vectors)
```

```python
# Create a scatter plot
plt.figure(figsize=(14, 10))
plt.scatter(pca_results[:, 0], pca_results[:, 1], c='blue')

# Annotate each point with its word label
for i, word in enumerate(words_to_plot):
    plt.annotate(word, xy=(pca_results[i, 0] + 0.03, pca_results[i, 1] + 0.03), fontsize=12)

plt.title("2D Visualization of Word Embeddings using PCA", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=12)
plt.ylabel("Principal Component 2", fontsize=12)
plt.grid(True)
plt.show()
```
