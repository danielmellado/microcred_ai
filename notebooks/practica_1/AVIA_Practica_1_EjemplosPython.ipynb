### 1. Ejemplo de lectura de datos desde fichero mediante la librería Numpy

Primero importamos la librería:

```python
import numpy as np
```

La librería contiene métodos para la lectura de datos de tipo numéricos desde un fichero de texto. Por ejemplo, mediente el método [genfromtxt](https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html "Referencia de genfromtxt"):

```python
Prostate = np.genfromtxt('/Prostate.csv', delimiter=',')
```

Con el atributo shape podemos saber el tamaño de la matriz de datos:

```python
Prostate.shape
```

También podemos visualizar el aspecto general de los datos imprimiendo simplemente el nombre de la variable. Como se puede observar la última columna contiene solo valores 0 y 1 indicando el valor de la clase de cada uno de los datos (filas):

```python
print(Prostate)
```

Podemos obtener subconjuntos de datos empleando índices de filas y columnas:

```python
Submatriz = Prostate[1:3,:3]
print(Submatriz)
```

También se pueden emplear índices negativos para empezar a numerar desde el final:

```python
Submatriz = Prostate[:,-3:]
print(Submatriz)
```

Se pueden realizar multitud de operaciones matriciales con los datos como, por ejemplo, una multiplicación de matrices (@):

```python
Nueva_matriz = Prostate @ Prostate.T
print(Nueva_matriz.shape)
```

Para poder visualizar un histograma con la distribución de los datos de cada una de las variables de los datos se puede emplear la librería de visualización [matplotlib](https://matplotlib.org). Por tanto, cargamos primero la librería:

```python
import matplotlib.pyplot as plt
```

Podemos mostrar el histograma de varias variables en una misma figura usando el método [hist](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html) en combinación con el método [subplot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html):

```python
fig, axes = plt.subplots(4, 4, figsize=(20,15))
for i in range(12):
    plt.subplot(3, 4, i+1)
    plt.hist(Prostate[:,i], bins=30)
    plt.title('Variable ' + str(i))
```

### 4. Ejemplos de normalización de datos

Importamos la librería de [pytorch](https://pytorch.org):

```python
import torch
```

Primero cargamos y convertimos los datos a tensores:

```python
Pacientes = pd.read_excel('/DatosSimples.xlsx', sheet_name='Censo')

Pacientes['Sexo'] = Pacientes['Sexo'].map({'H': 1, 'M': 0})

Pacientes_tensor = torch.tensor(Pacientes[['Edad', 'Altura']].values, dtype=torch.float32)
Pacientes.head()
```

Ahora normalizamos:

```python
data = torch.tensor(Pacientes.iloc[:,1:].values, dtype=torch.float32)
mean = data.mean(dim=0)
std = data.std(dim=0)
Pacientes_normalizados = (data - mean) / std
print("Pacientes normalizados:\n", Pacientes_normalizados)
print("Media = ", Pacientes_normalizados.mean(dim=0))
print("Desviación  = ", std)
```

### 5. Ejemplo de manejo de datos de imágenes:

Importamos la librería scikit-learn ya contiene un módulo ([datasets](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)) con utilidades para cargar conjuntos de datos populares en aprendizaje automático. También cuenta con algunos generadores de datos artificiales:

```python
plt.imshow(digitos.images[0]);
```
